```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
```

```python
# Cargamos la libreria
boston = load_boston()

print(boston.DESCR)
```

```python
x = np.array(boston.data[:,5])
y = np.array(boston.target)

print(x)
print(y)
```

```python
plt.scatter(x, y, alpha=0.3)
plt.show()
```

Formula de minimizar el error cuadratico medio (MCO): $\beta = (x^{T}x)^{-1}x^{T}y$

```python
#Aadimos columna de 1spara termino independiente
x_ones = np.array([np.ones(506),x]).T
print(x)
```

```python
b = np.linalg.inv(x_ones.T @ x_ones) @ x_ones.T @ y
```

```python
#De 4 a 9 (puntos del eje X), para y para cada -34 hay un 9.10...
plt.scatter(x, y, alpha=0.3)
plt.plot([4,9], [b[0] + b[1] * 4, b[0] + b[1] * 9], c = "red")
plt.show()
```
